##  第一题：

```python
year=1
while year<=5:
    fv=round(500*1.08**year,1)
    print(f"甲公司第{year}年末还款金额为{fv}万元")
    year= year+1
```

##  第二题：

```python
price=float(input("请输入产品单价："))
amount=float(input("请输入产品数量："))
day = int(input("请输入付款天数："))
if day<=10:
    sales = round(price*amount*0.98,1)
elif day <= 20:
    sales = round(price*amount*0.99,1)
else:
    sales = round(price*amount,1)
print(f"甲公司销售收入为{sales}元")
```
